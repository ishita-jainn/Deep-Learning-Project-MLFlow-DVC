## Ensure Annotations

```python
def prod(x:int, y:int)-> int:
    return x*y
```

```python
prod(2,"6")
```

```python
from ensure import ensure_annotations
@ensure_annotations
def prod(x:int, y:int)-> int:
    return x*y  
```

```python
prod(2,3)
```

```python
prod(2,"3")
```
